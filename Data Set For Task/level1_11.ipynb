```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

```

```python
data = pd.read_csv("3) Sentiment dataset.csv")
# data.head(5)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
le = LabelEncoder()
data["Platform_new"] = le.fit_transform(data["Platform"])
data.head(5)
```

```python
data["Sentiment"].unique()
```

```python
le = LabelEncoder()
le.fit(data["Sentiment"])
```

```python
data["Sentiment"] = le.transform(data["Sentiment"])
```

```python
data.head(5)
```

```python
ss = StandardScaler()
ss.fit(data[["Likes"]])
```

```python
data["Likes_new"] = pd.DataFrame(ss.transform(data[["Likes"]]), columns=["x"])
```

```python
data.info()
# data.columns
```

```python
x = data[["Year"]]
y = data["Day"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.6, random_state=42) 
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test, y_test)*100
```
